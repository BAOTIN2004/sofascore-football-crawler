```python
import time
import csv
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup

# ==== C·∫•u h√¨nh kho·∫£ng SBD c·∫ßn crawl ====
start_sbd = 80386
end_sbd = 99999

# ==== Kh·ªüi t·∫°o Selenium ====
driver = webdriver.Chrome()
driver.get("http://103.126.153.106/tracuu/index.html")

# ==== M·ªü file CSV ·ªü ch·∫ø ƒë·ªô append ====
output_file = "tuyensinh10.csv"
f = open(output_file, mode="a", newline="", encoding="utf-8-sig")
writer = None

empty_consecutive = 0
sbd_num = start_sbd

while sbd_num <= end_sbd:
    sbd = f"{sbd_num:06d}"
    try:
        print(f"\nüîÑ Tra c·ª©u SBD: {sbd}")

        # Nh·∫≠p v√† submit SBD
        input_box = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH,
                '/html/body/div/div[2]/div/form/table/tbody/tr/td/table[2]/tbody/tr[3]/td[3]/input'))
        )
        input_box.clear()
        input_box.send_keys(sbd)

        search_btn = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH,
                '/html/body/div/div[2]/div/form/table/tbody/tr/td/table[2]/tbody/tr[4]/td[3]/button'))
        )
        search_btn.click()

        # ƒê·ª£i b·∫£ng k·∫øt qu·∫£ hi·ªÉn th·ªã
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH,
                '/html/body/div/div[2]/div/table'))
        )
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")

        # L·∫•y c√°c d√≤ng d·ªØ li·ªáu
        rows = soup.find_all("tr", id="col_form")
        info = {}
        for row in rows:
            cells = row.find_all("td")
            if len(cells) >= 3:
                raw_label = " ".join(cells[0].stripped_strings)
                label = re.sub(r"\s+", " ", raw_label).strip()
                value = cells[2].get_text(strip=True)
                info[label] = value

        # Ki·ªÉm tra h·ª£p l·ªá: b·∫Øt bu·ªôc c√≥ "H·ªç v√† t√™n"
        ho_ten = info.get("H·ªç v√† t√™n", "").strip()
        if not ho_ten:
            empty_consecutive += 1
            print(f"‚ö†Ô∏è Kh√¥ng c√≥ d·ªØ li·ªáu (H·ªç v√† t√™n tr·ªëng). Li√™n ti·∫øp {empty_consecutive}/3")
        else:
            empty_consecutive = 0
            row_data = {
                "SBD": info.get("S·ªë b√°o danh", sbd),
                "H·ªç v√† t√™n": ho_ten,
                "Gi·ªõi t√≠nh": info.get("Gi·ªõi t√≠nh", ""),
                "Ng√†y sinh": info.get("Ng√†y sinh", ""),
                "Tr∆∞·ªùng TH/THCS": info.get("Tr∆∞·ªùng TH/THCS", ""),
                "Qu·∫≠n/Huy·ªán/TX": info.get("Qu·∫≠n/Huy·ªán/TX", ""),
                "Nguy·ªán v·ªçng 1": info.get("Nguy·ªán v·ªçng 1", ""),
                "Nguy·ªán v·ªçng 2": info.get("Nguy·ªán v·ªçng 2", ""),
                "Nguy·ªán v·ªçng 3": info.get("Nguy·ªán v·ªçng 3", ""),
                "ƒêi·ªÉm To√°n": info.get("ƒêi·ªÉm To√°n", ""),
                "ƒêi·ªÉm Ng·ªØ vƒÉn/Ti·∫øng Vi·ªát": info.get("ƒêi·ªÉm Ng·ªØ vƒÉn/Ti·∫øng Vi·ªát", ""),
                "ƒêi·ªÉm Ngo·∫°i Ng·ªØ": info.get("ƒêi·ªÉm Ngo·∫°i Ng·ªØ", ""),
                "T·ªïng x√©t ph·ªï th√¥ng/NTP": info.get("T·ªïng x√©t ph·ªï th√¥ng/NTP", ""),
                "K·∫øt qu·∫£": info.get("K·∫øt qu·∫£", "")
            }

            # Kh·ªüi t·∫°o writer n·∫øu c·∫ßn v√† ghi d√≤ng
            if writer is None:
                fieldnames = list(row_data.keys())
                writer = csv.DictWriter(f, fieldnames=fieldnames)
                if f.tell() == 0:
                    writer.writeheader()
            writer.writerow(row_data)
            print(f"‚úÖ Ghi SBD {sbd}: {row_data['H·ªç v√† t√™n']} - {row_data['K·∫øt qu·∫£']}")

        # Logic nh·∫£y +100 n·∫øu tr·ªëng 3 l·∫ßn li√™n ti·∫øp
        if empty_consecutive >= 3:
            print(f"‚è© Nh·∫£y l√™n SBD {sbd_num + 100:06d} do 3 SBD tr·ªëng li√™n ti·∫øp")
            sbd_num += 100
            empty_consecutive = 0
        else:
            sbd_num += 1

        time.sleep(0.5)  # Gi√£n nh·∫π gi·ªØa c√°c l·ªánh

    except Exception as err:
        print(f"‚ùå L·ªói t·∫°i SBD {sbd}: {err}")
        sbd_num += 1
        continue

# Ho√†n t·∫•t
driver.quit()
f.close()
print("\nüìÑ ƒê√£ crawl xong. K·∫øt qu·∫£ l∆∞u t·∫°i:", output_file)
```
