```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1= pd.read_csv('tuyensinh10TpHue.csv', encoding='utf-8')
```

```python
df2= pd.read_csv('tuyensinh10.csv', encoding='utf-8')
```

```python
df= pd.concat([df1, df2], ignore_index=True)
df
```

Kiểm tra và xóa các dòng giống hệt nhau (giữ lại bản ghi thứ nhất)

```python

duplicated_rows = df[df.duplicated()]
print(f"Số dòng trùng lặp hoàn toàn: {duplicated_rows.shape[0]}")
df = df.drop_duplicates()
```

các tên cột  viết thường, thay thế khoảng trắng

```python
from unidecode import unidecode

df.columns = [unidecode(col).lower().replace(" ", "_") for col in df.columns]
```

```python
df.head()
```

Kiểm tra dữ liệu thiếu

```python
df.isnull().sum()
```

Xử lý thiếu dữ liệu ở cột giới tính

```python
df['gioi_tinh'] = df['gioi_tinh'].replace('x', 'Nữ')
df['gioi_tinh'] = df['gioi_tinh'].replace('', 'Nam')  # thay chuỗi rỗng 
df['gioi_tinh'] = df['gioi_tinh'].fillna('Nam')       # thay NaN bằng Nam
```

Xóa các cột dữ liệu không có ý nghĩa

```python
df=df.drop(columns=['unnamed:_14', 'unnamed:_15', 'unnamed:_16', 'unnamed:_17', 
'unnamed:_18', 'unnamed:_19', 'unnamed:_20', 'unnamed:_21', 'unnamed:_22',
 'unnamed:_23', 'unnamed:_24', 'unnamed:_25', 'unnamed:_26'])
```

```python
df.head()
```

Chuyển đổi dữ liệu (từ string > float)

```python
cols_diem = ['diem_toan', 'diem_ngu_van/tieng_viet', 'diem_ngoai_ngu', 'tong_xet_pho_thong/ntp']

for col in cols_diem:
    df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
    df[col] = pd.to_numeric(df[col], errors='coerce')  # chuyển sang float, lỗi thì thành NaN
```

Các ô trống (không có điểm): không tham gia thi môn đó.

```python

df['diem_toan'] = df['diem_toan'].fillna(0)
df['diem_ngoai_ngu'] = df['diem_ngoai_ngu'].fillna(0)
df['diem_ngu_van/tieng_viet'] = df['diem_ngu_van/tieng_viet'].fillna(0)
df['tong_xet_pho_thong/ntp'] = df['tong_xet_pho_thong/ntp'].fillna(0)
```

Thêm cột điểm THCS

```python
df['diem_THCS']=(df['tong_xet_pho_thong/ntp']- df['diem_ngoai_ngu']- df['diem_ngu_van/tieng_viet']-df['diem_toan'])
```

```python
df['diem_THCS'] = df['diem_THCS'].fillna(0)
```
